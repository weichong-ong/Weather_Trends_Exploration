```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df_local = pd.read_csv('hamburg_data.csv',header=0,names=['year','local_avg_temp'])
df_local.set_index('year',inplace=True)
df_global = pd.read_csv('global_data.csv',header=0,names=['year','global_avg_temp'])
df_global.set_index('year',inplace=True)
```

```python
df = df_local.join(df_global, on='year')
```

## Create line charts

```python
fig, ax = plt.subplots(figsize=(15,6))
ax = df['local_avg_temp'].plot(legend=True,label='local')
ax = df['global_avg_temp'].plot(legend=True,label='global')
ax.set_ylabel('Average Temperature in °C',fontsize=12)
ax.set_xlabel('Year',fontsize=12)
ax.set_title('Weather Trend',fontsize=14)
plt.show()
```

## Create windows

```python
df['local_10_Year_MA'] = df['local_avg_temp'].rolling(window=10).mean()
df['global_10_Year_MA'] = df['global_avg_temp'].rolling(window=10).mean()
df['local_20_Year_MA'] = df['local_avg_temp'].rolling(window=20).mean()
df['global_20_Year_MA'] = df['global_avg_temp'].rolling(window=20).mean()
df['local_30_Year_MA'] = df['local_avg_temp'].rolling(window=30).mean()
df['global_30_Year_MA'] = df['global_avg_temp'].rolling(window=30).mean()
df['local_40_Year_MA'] = df['local_avg_temp'].rolling(window=40).mean()
df['global_40_Year_MA'] = df['global_avg_temp'].rolling(window=40).mean()
df['local_50_Year_MA'] = df['local_avg_temp'].rolling(window=50).mean()
df['global_50_Year_MA'] = df['global_avg_temp'].rolling(window=50).mean()
df.head(30)
```

## Plot the moving average 

```python
fig, ax = plt.subplots(nrows=5, ncols=1, figsize=(15,12))
ax[0].plot(df['local_10_Year_MA'],label='local_10-Year-MA')
ax[0].plot(df['global_10_Year_MA'],label='global_10-Year-MA')
ax[0].legend()
ax[0].grid('on', which='minor', axis='x')
ax[0].grid('off', which='major', axis='x')
ax[0].minorticks_on()

ax[1].plot(df['local_20_Year_MA'],label='local_20-Year-MA')
ax[1].plot(df['global_20_Year_MA'],label='global_20-Year-MA')
ax[1].legend()
ax[1].grid('on', which='minor', axis='x')
ax[1].grid('off', which='major', axis='x')
ax[1].minorticks_on()

ax[2].plot(df['local_30_Year_MA'],label='local_30-Year-MA')
ax[2].plot(df['global_30_Year_MA'],label='global_30-Year-MA')
ax[2].legend()
ax[2].grid('on', which='minor', axis='x')
ax[2].grid('off', which='major', axis='x')
ax[2].minorticks_on()

ax[3].plot(df['local_40_Year_MA'],label='local_40-Year-MA')
ax[3].plot(df['global_40_Year_MA'],label='global_40-Year-MA')
ax[3].legend()
ax[3].grid('on', which='minor', axis='x')
ax[3].grid('off', which='major', axis='x')
ax[3].minorticks_on()

ax[4].plot(df['local_50_Year_MA'],label='local_50-Year-MA')
ax[4].plot(df['global_50_Year_MA'],label='global_50-Year-MA')
ax[4].legend()
ax[4].grid('on', which='minor', axis='x')
ax[4].grid('off', which='major', axis='x')
ax[4].minorticks_on()

fig.suptitle('Weather trend exploration with 5 different moving average periods', ha='center', va='center', fontsize=15, y=0.92, fontweight='bold')
fig.yaxis_title='Average Temperature in °C'
fig.text(0.07, 0.5,'Average Temperature in °C', ha='center', va='center', rotation='vertical', fontsize=15)
fig.xaxis_title='Year'
fig.text(0.5, 0.07,'Year', ha='center', va='center', fontsize=15)

plt.show()
```

I plotted five line charts with different moving average periods to find out the best moving average period. From the figure above, we can see that the fluctuation between Year 1800-1850 started to disappeared with 30-Year moving average. Therefore, I decided to use 20-Year moving average to smooth out the lines, in order to observe the long term trend without losing too much information.

```python
fig, ax = plt.subplots(figsize=(15,6))
ax = df['local_20_Year_MA'].plot(legend=True,label='local')
ax = df['global_20_Year_MA'].plot(legend=True,label='global')
ax.grid('on', which='minor', axis='x')
ax.grid('off', which='major', axis='x')
plt.minorticks_on()
plt.legend(prop={'size': 12})
ax.set_ylabel('Average Temperature in °C',fontsize=12)
ax.set_xlabel('Year',fontsize=12)
ax.set_title('Weather Trend with 20-Year MA',fontsize=14, fontweight='bold')
plt.show()
```

```python
df['local_20_Year_MA'].corr(df['global_20_Year_MA'],method ='pearson')
```

**1. Is your city hotter or cooler on average compared to the global average? Has the difference been consistent over time?**  
Global average temperature (8.36°C) is higher on average compared to Hamburg’s average temperature (8.33°C). The difference has not been consistent as Hamburg was hotter on average before Year 1850.

**2. How do the changes in your city’s temperatures over time compare to the changes in the global average?**   
There was a bigger change in global average temperature between Year 1805 and Year 1825 compared to the changes in Hamburg’s average temperature. However, the changes in temperature were pretty similar for both global and local average temperature after Year 1850.

**3. What does the overall trend look like? Is the world getting hotter or cooler? Has the trend been consistent over the last few hundred years?**
- Over the last few hundred years, the world temperature has increased by approximately 1.5°C.
- Before Year 1900, there was only relatively small change in average temperature.
- After Year 1900, the world’s temperature has been increasing linearly.
- From Year 1980 onwards, there was an exponential upward trend. Both local and global temperature has increased by about 1°C in 30 years time.
